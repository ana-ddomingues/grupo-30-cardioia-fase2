```python
#!pip install rapidfuzz

import pandas as pd
from rapidfuzz import fuzz

from IPython.display import display, Markdown

# ====== 1. Crear archivos locales ======
with open("sintomas.txt", "w", encoding="utf-8") as f:
    f.write("""H√° dois dias estou com uma dor no peito que piora quando fa√ßo esfor√ßo f√≠sico.
Sinto cansa√ßo constante h√° uma semana, mesmo depois de descansar.
Tenho falta de ar ao subir escadas desde ontem.
Estou com dor de cabe√ßa forte h√° tr√™s dias e n√£o consigo me concentrar no trabalho.
Apresento febre alta e calafrios desde a madrugada.
Tenho dor abdominal intensa que come√ßou ap√≥s o almo√ßo de ontem.
Estou com dor de garganta e dificuldade para engolir h√° dois dias.
Tenho tontura frequente quando me levanto rapidamente.
Estou com tosse seca persistente h√° mais de uma semana.
Sinto dores nas articula√ß√µes e rigidez pela manh√£.""")

with open("mapa_sintomas.csv", "w", encoding="utf-8") as f:
    f.write("""Sintoma 1,Sintoma 2,Doen√ßa Associada
dor no peito,aperto no t√≥rax,Infarto
cansa√ßo constante,fadiga,Insufici√™ncia Card√≠aca
falta de ar,dificuldade para respirar,Angina
dor de cabe√ßa,enxaqueca,Enxaqueca
febre,calafrios,Infec√ß√£o Viral
dor abdominal,dor no est√¥mago,Gastrite
dor de garganta,dificuldade para engolir,Amigdalite
tontura,vertigem,Problemas de Press√£o
tosse seca,tosse persistente,Bronquite
dor nas articula√ß√µes,rigidez,Artrite""")

# 1.1 Exibi√ß√£o do conte√∫do dos arquivos (M√©todo Python)

display(Markdown('## Tabela: Relatos de Sintomas (`sintomas.txt`) üìù'))

# L√™ o arquivo de texto
with open("sintomas.txt", "r", encoding="utf-8") as f:
    # Cria uma lista de linhas, removendo espa√ßos e linhas vazias
    relatos = [linha.strip() for linha in f if linha.strip()]

# Cria o DataFrame
df_relatos = pd.DataFrame(relatos, columns=["Relato do Paciente"])
df_relatos.index.name = 'ID'
df_relatos.index = df_relatos.index + 1 # Inicia a contagem em 1

# Exibe a primeira tabela
display(df_relatos)

display(Markdown('## Tabela: Mapeamento de Sintomas (`mapa_sintomas.csv`) ü©∫'))

# L√™ o arquivo CSV diretamente para um DataFrame
df_mapa = pd.read_csv("mapa_sintomas.csv")

# Exibe a segunda tabela
display(df_mapa)



# ====== 2. Leer archivos ======
with open("sintomas.txt", "r", encoding="utf-8") as f:
    frases = f.readlines()

mapa = pd.read_csv("mapa_sintomas.csv")

# ====== 3. Funci√≥n para m√∫ltiples diagn√≥sticos ======
def sugerir_varios_diagnosticos(frase, mapa, limiar=70):
    frase_lower = frase.lower()
    diagnosticos_encontrados = []

    for _, row in mapa.iterrows():
        sintomas = [row["Sintoma 1"], row["Sintoma 2"]]
        for sintoma in sintomas:
            score = fuzz.partial_ratio(sintoma.lower(), frase_lower)
            if score >= limiar:
                # Guardar diagn√≥stico si a√∫n no est√° en la lista
                if row["Doen√ßa Associada"] not in diagnosticos_encontrados:
                    diagnosticos_encontrados.append(row["Doen√ßa Associada"])

    if diagnosticos_encontrados:
        return ", ".join(diagnosticos_encontrados)
    else:
        return "Sem diagn√≥stico definido"

# ====== 4. Testar diagn√≥sticos ======
resultados = []
for frase in frases:
    diag = sugerir_varios_diagnosticos(frase, mapa)
    resultados.append({"Frase": frase.strip(), "Diagn√≥sticos Sugeridos": diag})

df_resultados = pd.DataFrame(resultados)

display(Markdown('## Tabela:  Diagn√≥sticos com base na ontologia'))
import IPython.display as display
display.display(df_resultados)
```
